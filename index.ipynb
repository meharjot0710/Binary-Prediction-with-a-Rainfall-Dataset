```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
features=['pressure','maxtemp','temparature','mintemp','dewpoint','humidity','cloud','sunshine','winddirection','windspeed']
X=df[features]
Y=df['rainfall']
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.23, random_state=42)
X_train.head()
```

```python
model1=LogisticRegression()
model1.fit(X_train,y_train)

y_pred=model1.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
model2=RandomForestClassifier()
model2.fit(X_train,y_train)

y_pred=model2.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
model3 = xgb.XGBClassifier(objective='binary:logistic', n_estimators=150, learning_rate=0.2, max_depth=17, random_state=42)
model3.fit(X_train, y_train)

y_pred_xgb = model3.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("Classification Report:\n", classification_report(y_test, y_pred_xgb))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_xgb))
```

```python
model4=KNeighborsClassifier(10)
model4.fit(X_train,y_train)

y_pred=model4.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
